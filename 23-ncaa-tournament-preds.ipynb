```python
import statistics
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import neighbors
```

```python
df = pd.read_csv('../input/ncaa-tourney-training-data/TourneyFullData.csv')
df
```

```python
df = df.drop(columns = ['Team', 'Yr']);
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1]
```

```python
bay = linear_model.BayesianRidge()
bay.fit(x,y)
bay.coef_
```

```python
lars = linear_model.LassoLars(alpha=0.0001)
lars.fit(x,y)
lars.coef_
```

```python
n_neighbors = 20

for i, weights in enumerate(["uniform", "distance"]):
    knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
knn.fit(x,y)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
reg.coef_
```

```python
tourney = pd.read_csv('../input/ncaa-tourney-training-data/Fieldof682023.csv')
tourneyTeams = tourney.iloc[:,0]
tourneyTeams
```

```python
tourney = tourney.drop(columns = ['Team', 'Yr']);
```

```python
tourneyBay = bay.predict(tourney)
tourneyLars = lars.predict(tourney)
tourneyKnn = knn.predict(tourney)
tourneyReg = reg.predict(tourney)
tabletourney = pd.DataFrame({'teams':tourneyTeams, 'Bay':tourneyBay, 'Lars':tourneyLars, 'KNN':tourneyKnn, 'Linear':tourneyReg})
pd.set_option('display.max_rows', None)
tabletourney.sort_values("Bay")
```
