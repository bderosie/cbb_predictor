```python
# import all of the python modules/packages you'll need here
import matplotlib.pyplot as plt
import statistics
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn import linear_model
# ...
```

```python
df = pd.read_csv('../input/ncaa-tourney-training-data/TourneyPredictorTrainingData.csv')
df
```

```python
df = df.drop(columns = ['Team', 'Yr']);
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1]
X_train_raw, X_val_raw, y_train, y_val = train_test_split(x, y, test_size=0.15, random_state=68)
X_train_raw
```

# Bayesian Ridge Training

```python
bay = linear_model.BayesianRidge()
bay.fit(X_train_raw,y_train)
bay.coef_
```

```python
bayPreds = bay.predict(X_val_raw)
table = pd.DataFrame({'Actual':y_val, 'Predictions': bayPreds, 'Difference':(y_val - bayPreds)})
print(table)
print("Mean Squared Error: " + str(mean_squared_error(y_val, bayPreds)))
```

# Lasso Lars Training

```python
lars = linear_model.LassoLars(alpha=0.0001)
lars.fit(X_train_raw,y_train)
lars.coef_
```

```python
larsPreds = lars.predict(X_val_raw)
table1 = pd.DataFrame({'Actual':y_val, 'Predictions': larsPreds, 'Difference':(y_val - larsPreds)})
print(table1)
print("Mean Squared Error: " + str(mean_squared_error(y_val, larsPreds)))
```

# K Nearest Neighbors Training

```python
from sklearn import neighbors
n_neighbors = 20

for i, weights in enumerate(["uniform", "distance"]):
    knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
knn = knn.fit(X_train_raw,y_train)
```

```python
nnPreds = knn.predict(X_val_raw)
table2 = pd.DataFrame({'Actual':y_val, 'Predictions': nnPreds, 'Difference':(y_val - nnPreds)})
print(table2)
print("Mean Squared Error: " + str(mean_squared_error(y_val, nnPreds)))
```

# Linear Regression

```python
reg = linear_model.LinearRegression()
reg.fit(X_train_raw,y_train)
reg.coef_
```

```python
regPreds = reg.predict(X_val_raw)
table3 = pd.DataFrame({'Actual':y_val, 'Predictions': regPreds, 'Difference':(y_val - regPreds)})
print(table3)
print("Mean Squared Error: " + str(mean_squared_error(y_val, regPreds)))
```
